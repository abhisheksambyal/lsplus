```python
# https://waldyrious.net/viridis-palette-generator/
clrs = ['#fde725', '#86d549', '#2ab07f', '#fec488', '#ed6925', '#482173',] # customized
# clrs = ['#EE7733', '#0077BB', '#33BBEE', '#EE3377', '#CC3311', '#009988', ][::-1] # https://personal.sron.nl/~pault/
# clrs = ['#fde725', '#35b779', '#31688e', '#440154'] # virdis
# clrs = ['#fcffa4', '#ed6925', '#781c6d', '#000004'] # inferno
# clrs = ['#fcfdbf', '#f1605d', '#721f81', '#000004'] # magma
# clrs = ['#f0f921', '#ed7953', '#9c179e', '#0d0887'] # plasma

fracs_retained = np.log(np.arange(200 + 1)[1:])
fracs_retained /= np.amax(fracs_retained)
    
plt.rcParams.update({'font.size': 13})
plt.figure(figsize=(16,5))
fslegend = 11
bbox_anchor = (0.5, -.35)

# Dataset1
m_name = 'Resnet34'  # run:1
DIR_D1=f'd1_code/retention_curve_files'
y_baseline = np.loadtxt(f'{DIR_D1}/{m_name}_baseline.txt', delimiter=",")    
y_LS = np.loadtxt(f'{DIR_D1}/{m_name}_LS.txt', delimiter=",")    
y_FL = np.loadtxt(f'{DIR_D1}/{m_name}_FL.txt', delimiter=",")
y_dca = np.loadtxt(f'{DIR_D1}/{m_name}_dca.txt', delimiter=",")    
y_mdca = np.loadtxt(f'{DIR_D1}/{m_name}_mdca.txt', delimiter=",")
y_ours = np.loadtxt(f'{DIR_D1}/{m_name}_ours.txt', delimiter=",")    
plt.subplot(141)
plt.plot(fracs_retained, y_baseline*100, label=f'HL: {np.trapz(y_baseline, x = fracs_retained)*100:.2f}', color=clrs[0])
plt.plot(fracs_retained, y_LS*100, label=f'LS: {np.trapz(y_LS, x = fracs_retained)*100:.2f}', color=clrs[1])
plt.plot(fracs_retained, y_FL*100, label=f'FL: {np.trapz(y_FL, x = fracs_retained)*100:.2f}', color=clrs[2])
plt.plot(fracs_retained, y_dca*100, label=f'DCA: {np.trapz(y_dca, x = fracs_retained)*100:.2f}', color=clrs[3])
plt.plot(fracs_retained, y_mdca*100, label=f'MDCA: {np.trapz(y_mdca, x = fracs_retained)*100:.2f}', color=clrs[4])
plt.plot(fracs_retained, y_ours*100, label=f'$Ours$: {np.trapz(y_ours, x = fracs_retained)*100:.2f}', color=clrs[5])
plt.xlabel('Retention Fraction')
plt.ylabel('Accuracy')
plt.title(f'Chaoyang | ResNet34')
plt.ylim(78, 102)
plt.yticks(np.arange(80, 101, 5))
plt.grid()
plt.legend(fontsize=fslegend, edgecolor='black', loc='center', bbox_to_anchor=bbox_anchor, ncol=2)

# Dataset2
m_name = 'Resnet34' # run:2
DIR_MHIST=f'code_mhist/retention_curve_files'
y_baseline = np.loadtxt(f'{DIR_MHIST}/{m_name}_baseline.txt', delimiter=",")    
y_LS = np.loadtxt(f'{DIR_MHIST}/{m_name}_LS.txt', delimiter=",")    
y_FL = np.loadtxt(f'{DIR_MHIST}/{m_name}_FL.txt', delimiter=",")    
y_dca = np.loadtxt(f'{DIR_MHIST}/{m_name}_dca.txt', delimiter=",")    
y_mdca = np.loadtxt(f'{DIR_MHIST}/{m_name}_mdca.txt', delimiter=",")    
y_ours = np.loadtxt(f'{DIR_MHIST}/{m_name}_ours.txt', delimiter=",")  
plt.subplot(142)
plt.plot(fracs_retained, y_baseline*100, label=f'HL: {np.trapz(y_baseline, x = fracs_retained)*100:.2f}', color=clrs[0])
plt.plot(fracs_retained, y_LS*100, label=f'LS: {np.trapz(y_LS, x = fracs_retained)*100:.2f}', color=clrs[1])
plt.plot(fracs_retained, y_FL*100, label=f'FL: {np.trapz(y_FL, x = fracs_retained)*100:.2f}', color=clrs[2])
plt.plot(fracs_retained, y_dca*100, label=f'DCA: {np.trapz(y_dca, x = fracs_retained)*100:.2f}', color=clrs[3])
plt.plot(fracs_retained, y_mdca*100, label=f'MDCA: {np.trapz(y_mdca, x = fracs_retained)*100:.2f}', color=clrs[4])
plt.plot(fracs_retained, y_ours*100, label=f'$Ours$: {np.trapz(y_ours, x = fracs_retained)*100:.2f}', color=clrs[5])
plt.xlabel('Retention Fraction')
plt.ylabel('Accuracy')
plt.title(f'MHIST | ResNet34')
plt.ylim(78, 102)
plt.yticks(np.arange(80, 101, 5))
plt.grid()
plt.legend(fontsize=fslegend, edgecolor='black', loc='center', bbox_to_anchor=bbox_anchor, ncol=2)


# Dataset3
m_name = 'Resnet34' # run:2
DIR_MHIST=f'd3_skin_code/retention_curve_files'
y_baseline = np.loadtxt(f'{DIR_MHIST}/{m_name}_baseline.txt', delimiter=",")    
y_LS = np.loadtxt(f'{DIR_MHIST}/{m_name}_LS.txt', delimiter=",")    
y_FL = np.loadtxt(f'{DIR_MHIST}/{m_name}_FL.txt', delimiter=",")    
y_dca = np.loadtxt(f'{DIR_MHIST}/{m_name}_dca.txt', delimiter=",")    
y_mdca = np.loadtxt(f'{DIR_MHIST}/{m_name}_mdca.txt', delimiter=",")    
y_ours = np.loadtxt(f'{DIR_MHIST}/{m_name}_ours.txt', delimiter=",")  
plt.subplot(143)
plt.plot(fracs_retained, y_baseline*100, label=f'HL: {np.trapz(y_baseline, x = fracs_retained)*100:.2f}', color=clrs[0])
plt.plot(fracs_retained, y_LS*100, label=f'LS: {np.trapz(y_LS, x = fracs_retained)*100:.2f}', color=clrs[1])
plt.plot(fracs_retained, y_FL*100, label=f'FL: {np.trapz(y_FL, x = fracs_retained)*100:.2f}', color=clrs[2])
plt.plot(fracs_retained, y_dca*100, label=f'DCA: {np.trapz(y_dca, x = fracs_retained)*100:.2f}', color=clrs[3])
plt.plot(fracs_retained, y_mdca*100, label=f'MDCA: {np.trapz(y_mdca, x = fracs_retained)*100:.2f}', color=clrs[4])
plt.plot(fracs_retained, y_ours*100, label=f'$Ours$: {np.trapz(y_ours, x = fracs_retained)*100:.2f}', color=clrs[5])
plt.xlabel('Retention Fraction')
plt.ylabel('Accuracy')
plt.title(f'ISIC | ResNet34')
plt.ylim(68, 103)
plt.yticks(np.arange(70, 101, 5))
plt.grid()
plt.legend(fontsize=fslegend, edgecolor='black', loc='center', bbox_to_anchor=bbox_anchor, ncol=2)



plt.tight_layout()
plt.savefig('r34_retention_curves.png', dpi=300, bbox_inches='tight')
plt.show()
```
